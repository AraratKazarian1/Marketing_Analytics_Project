### Example of using the ComboGenius package

```python
# Import the necessary functions from the package
from combogenius.database.schema import *
from combogenius.database.sql_interactions import SqlHandler
from combogenius.logger.logger import CustomFormatter
from combogenius.models.make_combos import combos
import pandas as pd
```

```python
# Create the database and load the data
Inst  = SqlHandler('database', 'checks')
Inst1 = SqlHandler('database', 'companies')
Inst2 = SqlHandler('database', 'price_list')

data=pd.read_csv('sample_data/data.csv')
companies = pd.read_csv('sample_data/companies.csv')
price_list = pd.read_csv('sample_data/price_list.csv')
Inst.insert_many(data)

Inst1.insert_many(companies)
Inst2.insert_many(price_list)

Inst.close_cnxn()
```

**The ComboGenius package serves a variety of purposes, one of which is assisting in combo creation. It works by analyzing data on checks and prices to identify the most sought-after combinations of products. By calculating both the price and frequency of these combinations, it helps businesses optimize their offerings to meet customer demand more effectively.**

```python
# Make new combos
m = combos()
f = m.make_combos(5)
print(f)
```

```python
# Make visualizations
m.visualize_most_frequent_combos()
m.visualize_expensive_combos()
m.visualize_cheap_combos()
```

**Another key feature of the ComboGenius package is its capability to improve communication with B2B customers by making it easier to share new combo offerings. Utilizing an API, it streamlines the process of sending emails to notify clients about these updated combinations, thereby enhancing overall communication efficiency.**
